## Columns Manipulation
    - Create new cloumn
    - Rename column
    - Drop Column

```python
df_new=df.copy()
```

```python
df_new.head(3)
```

```python
# Create New Column
# np.absolute untuk membuat niali absolut tanpa -
df_new['selisih'] = np.absolute(df_new['pts'] - df_new['opp_pts'])
```

```python
df_new.head(2)
```

```python
# Rename
df_new.rename(columns={'game_location' : 'loc',
            'game_result' : 'result'})
```

```python
df_new.rename(columns={'game_location' : 'loc',
            'game_result' : 'result'}, inplace= True )
```

```python
df.head(3)
```

```python
# Drop Column
# Axis 0 = Baris
# Axis 1 = Kolom
df_new = df_new.drop(['elo_i', 'elo_n'], axis=1)
```

```python
df_new.head(3)
```

```python
# creating new column based on existing information using loop

# siapkan list kosong
# buat program loop -> selisih pts dengan opp_pts
# simpan hasil per baris ke list kosong
# convert list yang sudah terisi menjadi kolom

hasil = []

# Menggunakan iloc
# df_new.iloc[:, 10] index kolom pts
# df_new.iloc[:, 14] index kolom opp_pts
# atau bisa menggunakan loc
# df_new.iloc[:, pts] kolom pts
# df_new.iloc[:, opp_pts] kolom opp pts

for i in range(len(df_new)):
    selisih = df_new.iloc[i,10]- df_new.loc[i, 'opp_pts']
    hasil.append(selisih)
df_new['menggunakan loop'] = np.abs(hasil)
```
